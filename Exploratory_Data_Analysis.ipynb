GroupBy in Python


```python
df_1 = df[['drive-wheels','price']]
df_1.head()
```

```python
df_1_grp = df_1.groupby(['drive-wheels'], as_index = False).mean()
df_1_grp
```

```python
df[['body-style']].value_counts()
```

```python
df_price_body = df[['drive-wheels','body-style','price']]
df_price_body_grp = df_price_body.groupby(['drive-wheels','body-style'], as_index=False).mean()
df_price_body_grp.head()
```

```python
df_pivot = df_price_body_grp.pivot(index='drive-wheels',columns='body-style')
df_pivot
```

Heat Map Plot

```python
#from above pivot 
plt.pcolor(df_pivot,cmap='RdBu')
plt.colorbar()
plt.show()
```

Pearson Correlation - a) Corr coeff & b)p - value

```python
#removing rows with NaN or infinite values
df_cleaned = df[['horsepower','price']].dropna()
df_cleaned = df_cleaned[np.isfinite(df_cleaned).all(1)]
#calculating pearson's correlation
pearsonr_coef, p_value = stats.pearsonr(df_cleaned['horsepower'],df_cleaned['price'])
pearsonr_coef, p_value
```

```python
#determining correlation for numeric data only
df_numeric = df.select_dtypes(include=[float,int])
corr_matrix = df_numeric.corr()
corr_matrix
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(corr_matrix,annot=True,cmap='coolwarm',fmt='.2f',linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
#alternative correlational heatmap approach
df_heat = df.corr(numeric_only=all)
plt.figure(figsize=(10,6))
sns.heatmap(df_heat,annot=True,cmap='coolwarm',fmt='.2f', linewidths=0.5)
plt.title('Correlational Heatmap Alternative')
plt.show()
```
